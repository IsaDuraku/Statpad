```python
matches["Date"] = pd.to_datetime(matches["Date"], format="%d/%m/%Y")
```

```python
matches["home_code"] = matches["HomeTeam"].astype("category").cat.codes
matches["away_code"] = matches["AwayTeam"].astype("category").cat.codes
```

```python
matches["hour"] = matches["Time"].str.replace(":.+", "",regex =True).astype("int")
```

```python
matches
```

```python
matches["day_code"] = matches["Date"].dt.dayofweek
```

```python
matches
```

```python
matches["home_win"] = (matches["FTR"] == "H").astype("int")
matches["away_win"] = (matches["FTR"] == "A").astype("int")
unique_teams = pd.concat([matches["home_code"], matches["away_code"]]).unique()

# Initialize the "win" column with 0
matches["win"] = 0

# Update the "win" column based on the desired logic
for team in unique_teams:
    matches.loc[((matches["home_code"] == team) & (matches["home_win"] == 1)) | ((matches["away_code"] == team) & (matches["away_win"] == 1)), "win"] = 1
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
train = matches[matches["Date"] < '2023-01-01']
```

```python
test = matches[matches["Date"] > '2023-01-01']
```

```python
predictors = ["home_code","away_code","hour","day_code","home_win","away_win"]
```

```python
rf = RandomForestClassifier(n_estimators=50,min_samples_split=10, random_state=1)
```

```python
rf.fit(train[predictors], train["target"])
```

```python
preds = rf.predict(test[predictors])
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc = accuracy_score(test["target"], preds)
```
